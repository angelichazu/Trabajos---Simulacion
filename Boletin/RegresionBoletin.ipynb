```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime,timedelta
from sklearn.model_selection import train_test_split
import datetime as dt
from sklearn.linear_model import LinearRegression  
import altair as alt
import altair_viewer
from sklearn.preprocessing import PolynomialFeatures
```

<h2><span style="color: #00ff00;"> Importaci&oacute;n de datos</span></h2>
<p>Se importa el dataset que se ultizará para obtener todos los datos necesarios</p>

```python
df = pd.read_csv('./Datos.csv',encoding='utf-8')
df = df.fillna(0)
df.head()
```

<h2>
	<span style="color: #00ff00;">
		 Preprocesamiento
	</span>
</h2>

Para una fácil lectura de los datos, intercambiamos las filas por columnasdata.describe().round(3)

```python
data = df.T
data.columns = df['Country Name']
data = data[4:]
data.head()
```

<h2>
	<span style="color: #00ff00;">
		 Modelo Matemático
	</span>
</h2>

Para el modelo matemático utilizaremos una regresión lineal, el cual nos permitirá predecir valores del PIB en futuros años, para ello utilizaremos como entrada el año y como salida el PIB

```python
alt.Chart(data).mark_bar(color='orange').encode(
    x=alt.X('Year',sort=alt.EncodingSortField(field="id", op="count", order='ascending')),
    y='Ecuador'
    ).interactive()
```

<h2>Dividimos los datos que utilizaremos en el modelo matemático</h2>

```python
X=data['Year'].values.reshape(-1,1)
y=data['Ecuador']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=42)
```

<h2>Definimos nuestro modelo matematico, y lo entrenamos</h2>

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

```python
print('Precisión: ', linear_regressor.score(X_train, y_train))
```

<h2>La precisión de nuestro es muy baja, por ende definiremos un modelo polinomial</h2>

```python
poly_reg1 =  PolynomialFeatures(degree = 4)
poly_reg =  PolynomialFeatures(degree = 6)
X_poly = poly_reg1.fit_transform(X_train)
lin_reg_1 = LinearRegression()
lin_reg_1.fit(X_poly, y_train)
```

```python
print('Precisión: ', lin_reg_1.score(X_poly, y_train))
```

<h2>Una vez mejorada nuestra precisión, podemos predecir futuros valores del PIB en nuestro país</h2>

```python
X_poly2 = poly_reg1.fit_transform(X_test)
predi=lin_reg_1.predict(X_poly2)
print(predi)
```
